```python
import dill
import codecs
import pyautogui
import time
import pyperclip
import numpy as np
from pyperiscope import Scope

automation_payload = {}
```

# 0. Multistep collection of agents and emails

```python
# comment: Automated step generated with pyPeriscope V2 
payload = ''''''
step = Scope(saved_dict=dill.loads(codecs.decode(payload.encode(), 'base64')))
step.render_preview()
```

```python
payload = ''''''
step_check = Scope(saved_dict=dill.loads(codecs.decode(payload.encode(), 'base64')))
step_check.render_preview()
```

```python
payload = ''''''
step_agent = Scope(saved_dict=dill.loads(codecs.decode(payload.encode(), 'base64')))
step_agent.render_preview()
```

```python
payload = ''''''
step_at = Scope(saved_dict=dill.loads(codecs.decode(payload.encode(), 'base64')))
step_at.render_preview()
```

```python
automation_agents = []
automation_emails = []
last_header = ""

def find_in_list(in_list, val):
    ret_val = "NA"
    if len(in_list) == 1:
        return in_list[0][0]
        
    for item in in_list:
        if item[1] > val:
            break
        ret_val = item[0]
    return ret_val  

for i in range(10):
    cropped = pyautogui.screenshot()
    cropped = cropped.crop(((0,0,cropped.size[0]/3, cropped.size[1])))
    limit = -1
    header_list = []
    
    #find last header (Publicist,...) and limit the until there
    try:
        step_check.find(confidence=0.99, input_image=cropped)
        started = False
        at_end = -1
        for j in range(len(step_check.found_locations)):
            pyautogui.click(step_check.get_locations()[j], clicks=3)
            time.sleep(1)
            pyautogui.hotkey('ctrl','c')
            header = pyperclip.paste().strip()
            header_list.append([header,step_check.get_locations()[j][1]])
            if any(word in header.lower() for word in ["publicist","legal","commercial","voice","literary"]):
                limit = step_check.get_locations()[j][1]
                break
                
            if started and any(word in header.lower() for word in ["company","direct"]):
                at_end = step_check.get_locations()[j][1]
            if any(word in header.lower() for word in ["company","direct"]):
                at_end = cropped.size[1]
                started = True
    except Exception as e:
        print(e)

    #add the previous page last as the 1st 
    if last_header!="":
        header_list.insert(0, [last_header[0],0])
        
    #if the end header (Publicist,...) is found crop the image from bottom to limit hits
    if limit>0:
        subcrop = cropped.crop((0,0,cropped.size[0], limit))
    else:
        subcrop = cropped
        
    #find agent links in the remaining image
    try:
        step_agent.find(confidence=0.96, input_image=subcrop)
        for j in range(len(step_agent.found_locations)):
            # copy the name by 3x click on left of the link
            pyautogui.click((step_agent.get_locations()[j][0]+200,step_agent.get_locations()[j][1]), clicks=3)
            time.sleep(1)
            pyautogui.hotkey('ctrl','c')
            agent = pyperclip.paste().strip()
            print("Agent: " + agent)
            # copy the IMDB link for the agent
            pyautogui.click(step_agent.get_locations()[j], button="right")
            for y in range(5):
                pyautogui.hotkey("down")
            pyautogui.hotkey("enter")
            time.sleep(1)
            link = pyperclip.paste().strip().split("?")[0]
            print("Agent imdb: " + link)
            role = find_in_list(header_list, int(step_agent.get_locations()[j][1]))
            print("Agent role: " + role)
            pyautogui.click((step_agent.get_locations()[j][0]-50, step_agent.get_locations()[j][1]))
            time.sleep(1)
            automation_agents.append({"name":agent,"url":link, "role": role})
    except Exception as e:
        print(e)
        
    #find emails in the ramaining image
    email = "NA"
    try:
        step_at.find(confidence=0.72, input_image=subcrop)
        for j in range(len(step_at.found_locations)):
            if at_end > 0 and step_at.get_locations()[j][1] > at_end:
                break

            role = find_in_list(header_list, int(step_at.get_locations()[j][1]))
            if not role in ["Company", "Direct Contact"]:
                break
                
            print("email role: " + role)
            
            pyautogui.click(step_at.get_locations()[j], button="right")
            for y in range(4):
                pyautogui.hotkey("down")
            pyautogui.hotkey("enter")
            time.sleep(1)
            email = pyperclip.paste().strip()
            print("email: " + email)
            
            automation_emails.append({"email":email,"role":role})
    except Exception as e:
        print(e)
        
    #if it is not the end scroll down a page
    try:
        if limit<0:
            step.find(confidence=0.99, input_image=cropped)
        break
    except:
        pyautogui.scroll(-1170)
        last_header=header_list[-1]
        print("cont...")

automation_payload["emails"] = automation_emails
automation_payload["agents"] = automation_agents
pyautogui.hotkey('home')
```
