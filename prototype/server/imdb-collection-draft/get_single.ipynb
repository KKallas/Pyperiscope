```python
import dill
import codecs
import pyautogui
import time
import pyperclip
from pyperiscope import Scope
```

# 0. get the name

```python
# comment: Automated step generated with pyPeriscope V2 
payload = ''''''
step = Scope(saved_dict=dill.loads(codecs.decode(payload.encode(), 'base64')))
step.render_preview()
```

```python
step.find(confidence=0.99)
```

```python
data = []
for i in range(60):
    try:
        step.find(confidence=0.99)
        pyperclip.copy("--")
        # get the name
        pyautogui.moveTo((step.get_locations()[0][0]+13,step.get_locations()[0][1]-30))
        pyautogui.drag(0,-30,0.25, button="left")
        pyautogui.hotkey("ctrl","c")
        name_val = pyperclip.paste()
        # if the title is 2 lines
        if name_val=="--":
            pyautogui.moveTo((step.get_locations()[0][0]+13,step.get_locations()[0][1]+20))
            pyautogui.drag(0,-60,0.25, button="left")
            pyautogui.hotkey("ctrl","c")
            name_val = pyperclip.paste()
        name_val = name_val[:int(len(name_val)/2)-1]
        # for reasons beyond... it is 2x so need to make it half the length
        print(name_val)
    
        # get direct link
        pyautogui.click((step.get_locations()[0][0]+40,step.get_locations()[0][1]-60), button="right")
        time.sleep(0.5)
        for i in range(5):
            pyautogui.hotkey("down")
        pyautogui.hotkey("enter")
        url_val = pyperclip.paste().split("?")[0]
        print(url_val)
        data.append({"imdb_id":url_val.split("/")[-1],"name":name_val,"URL":url_val})
        # mark line done
        pyautogui.click(step.get_locations()[0])

        # scroll down about one film
        pyautogui.scroll(-185)
    except:
        break
```

```python
import pandas as pd

def update_films(new_films, store_path='film_database.h5'):
    with pd.HDFStore(store_path) as store:
        if 'films' in store:
            existing_films = store['films']
            
            # Find only the new films (not in existing imdb_ids)
            existing_ids = set(existing_films['imdb_id'])
            new_films = new_films[~new_films['imdb_id'].isin(existing_ids)]
            
            if not new_films.empty:
                # Append new films to existing ones
                updated_films = pd.concat([existing_films, new_films], ignore_index=True)
                store['films'] = updated_films
                print(f"Added {len(new_films)} new films")
            else:
                print("No new films to add")
        else:
            # If no existing films, create new table
            store['films'] = new_films
            print(f"Created new films table with {len(new_films)} entries")

new_films = pd.DataFrame(data)
update_films(new_films,"film_database.h5")
```
